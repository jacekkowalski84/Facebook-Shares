```python
import pandas as pd
import plotly.graph_objects as go

def remove_dollar(value:str)->float:
    return float(value[1:])

df= pd.read_csv ('data/fb.csv',
                parse_dates=['Date'],
                index_col = 'Date',
                engine = 'python'
                ).rename(columns={'Close/Last': 'Close_Last'}
                        ).assign(
                                Close_Last = lambda x: x['Close_Last'].map(remove_dollar),
                                Open = lambda x: x['Open'].map(remove_dollar),
                                High = lambda x: x['High'].map(remove_dollar),
                                Low = lambda x: x['Low'].map(remove_dollar)
                                ).sort_values(by=['Date'], axis=0, ascending=True
                                             ).assign (Close = lambda x: x['Open'].shift(1, axis=0)
                                                      ).assign (Open_Avarage = lambda x: x['Open'].rolling(14).mean()
                                                               )
df.index = df.index.strftime('%Y/%m/%d')
df
```

```python
df ['2019':'2020 1'].plot (
    y=['Open', 'Open_Avarage', 'Volume'],
    figsize=(15, 5),
    secondary_y=('Volume'),
    grid=True    
);
```

```python
fig = go.Figure(data=
                [go.Candlestick(x=df.index,
                open=df['Open'],
                high=df['High'],
                low=df['Low'],
                close=df['Close'])]
               )

fig 
```
